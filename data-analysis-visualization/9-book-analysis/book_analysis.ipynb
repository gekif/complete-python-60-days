# Load the data

```python
with open("miracle_in_the_andes.txt", "r", encoding="utf-8") as file:
    book = file.read()
```

```python
type(book)
```

# How Many Chapters

### With String Methods

```python
book.count("Chapter")
```

### With Regex

```python
import re
```

```python
pattern = re.compile("Chapter [0-9]")
pattern
```

```python
findings = re.findall(pattern, book)
findings
```

```python
len(findings)
```
